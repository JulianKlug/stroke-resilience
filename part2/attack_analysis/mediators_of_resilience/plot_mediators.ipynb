```python
import pandas as pd
import seaborn as sns
import scipy as sp
```

```python
data_path = '/Users/jk1/temp/resilience_part2/atttacks/data_globalEff_mediators_P2.sav'
```

```python
data = pd.read_spss(data_path)
```

```python
data.head()
```

```python
mediators = data.columns.values[2:] 
mediators
```

```python
selected_mediators = ['lesion_side', 'wmd', 'wmgm']
```

```python
# for each mediator, plot the effect of mediator on globEff (correlation)

for mediator in selected_mediators:
    g = sns.lmplot(x=mediator, y='globEff', data=data)
    ax = g.ax
    # define titles and axes labels
    ax.set_xlabel(mediator)
    ax.set_ylabel('Delta Global Efficiency')

    
```
